```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("fivethirtyeight")
```

```python
df = pd.read_csv("population-lang-sex-age.csv")
df.head(10)
```

```python
"""
TASKS:
 - population based on language
 - population based on gender
 - population based on district
"""
```

```python
language_result = df.groupby("Language")
languages = [x for x,df in language_result]
population = language_result.sum()["Value"]
plt.figure(figsize=(18,10))
plt.title("Population Based on Language (11/12)")
plt.bar(languages, population)
plt.xlabel("Language")
plt.ylabel("Population")
plt.xticks(languages, rotation="vertical")
plt.show()
```

```python
gender = df.groupby("Gender")
genders = [x for x,df in gender]
population = gender.sum()["Value"]
plt.figure(figsize=(18,10))
plt.title("Population based on gender (11/12)")
plt.bar(genders, population)
plt.show()
```

```python
district = df.groupby("District")
districts = [x for x,df in district]
value = district.sum()["Value"]

plt.figure(figsize=(18,10))
plt.title("Population based on District")
plt.bar(districts, value, color="r")
plt.xticks(districts, rotation="vertical")
plt.xlabel("Districts")
plt.ylabel("Population")
plt.show()
```
